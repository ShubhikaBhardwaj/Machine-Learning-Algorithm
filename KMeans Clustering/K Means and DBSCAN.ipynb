```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles,make_moons
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
X,Y= make_moons(n_samples=200,noise=0.10)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
km=KMeans(n_clusters=2)
km.fit(X)
```

```python
label=km.labels_
centers=km.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],c=label)
plt.scatter(centers[:,0],centers[:,1],color="red",marker="*")
plt.show()
```

### K-Means Clustering algorithm doesn't work well with complex datasets like moon shaped data 

# DB Scan Algorithm
* DBSCAN is preferred in case of complex datasets

* #### Density based Spatial clustering of Applications with Noise
* Idea-> The cluster is a high density area surrounded by low density area.
* Starts by exploring a small area, if the density is "good enough" it is comsidered as a part of the cluster
* And explores neighbours and so on, so it groups regions based upn their density!

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
#DBSCAN?

#Eps-> choosen based on the scale of your dataset
#Min_Samples -> mostly 5
```

* Eps ->The maximum distance between two samples for one to be considered
    as in the neighborhood of the other.
* Min_Samples ->The number of samples (or total weight) in a neighborhood for a point
    to be considered as a core point. This includes the point itself.


```python
dbs=DBSCAN(eps=0.1,min_samples=5)
dbs.fit(X)

```

```python
Ypred=dbs.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=Ypred)
plt.show()
```

### Try to improve by changing the eps and no of samples

```python
dbs=DBSCAN(eps=0.2,min_samples=5)
dbs.fit(X)
Ypred=dbs.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=Ypred)
plt.show()
```
