## Data Augmentation
- Deep Learning Models are "Data Hungry"
- Small Datasets lead to overfitting most of the times
- So lets generate more data by augmentating the training samples

```python
img_path = "C:/Users/sumeet/Desktop/machine-learning-online-2018/Datasets/Images/images/horses/horse-100.jpg"
```

```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
img = image.load_img(img_path,target_size=(100,100))
```

```python
x = image.img_to_array(img)/255.0
```

```python
plt.imshow(x)
plt.axis("off")
plt.show()
```

```python
x = x.reshape((1,100,100,3))
```

```python
datagen = image.ImageDataGenerator(
    rotation_range = 30,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = "nearest"
)
```

```python
i = 0
for batch in datagen.flow(x,batch_size=1):
    plt.figure()
    imgplot = plt.imshow(image.img_to_array(batch[0]))
    i+=1
    if i==10:
        break
    plt.axis("off")
    plt.show()
    
    
```
