# CNN

* CNN Layers are made up of filters

* When the feature is present in part of an image, the convolution operation between the filter and that part of the image results in a real number with a high value.
* If the feature is not present, the resulting value is low.

```python

import numpy as np
from keras.datasets import mnist
import matplotlib.pyplot as plt
import cv2
```

```python
def drawImg(img,title="Image"):
    plt.imshow(img,cmap='gray')
    plt.axis("off")
    plt.style.use("seaborn")
    plt.title(title+str(img.shape))
    plt.show()
```

```python
img_ = cv2.imread("./images/cody2.jpg")
img_ = cv2.resize(img_,(100,100))
print(img_.shape)
img = cv2.cvtColor(img_,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
drawImg(img,"RGB")
drawImg(img_gray,"Gray Scale")
```

```python
def convolution(img,img_filter):
    W = img.shape[0]
    H = img.shape[1]
    
    F = img_filter.shape[0]
    new_img = np.zeros((W-F+1,H-F+1))
    
    for row in range(W-F+1):
        for col in range(H-F+1):
            for i in range(F):
                for j in range(F):
                    new_img[row][col] += img[row+i][col+j]*img_filter[i][j]
                if new_img[row][col] > 255:
                    new_img[row][col] = 255
                elif new_img[row][col] < 0:
                    new_img[row][col] = 0
                    
    return new_img
```

```python

blur_filter = np.ones((3,3))/9.0
output1 = convolution(img_gray,blur_filter)
```

```python

edge_filter = np.array([[1,0,-1],
                        [1,0,-1],
                        [1,0,-1]])
output2 = convolution(img_gray,edge_filter)
```

```python
drawImg(img_gray,"Original")
drawImg(output1,"Blur Filter")
drawImg(output2,"Edge Detector")
```

* So Filters can highlight "Features" that can be useful to classify images. 
* Color is not always important, features like edges help in detecting objects!
